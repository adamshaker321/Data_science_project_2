# LIBRARIES

```python
import numpy as np
import pandas as pd
import re
```

# READING DATASET

```python
df = pd.read_excel(r"C:\Users\Me\Downloads\DS_project_1\abe_books_web.xlsx",sheet_name="abe_books_web")
df.shape
df.index
df.info()
```

# DATA CLEANING ( 1_ TEXT CLEANING)

```python
df['cover_type'].head()

# handling cover_type column
df['cover_type']=df['cover_type'].str.replace('Used - ','')

# handling shipping_price column
df['shipping_price']=df['shipping_price'].str.extract(r'(\d+.\d+)').astype(float)
df['shipping_price']

# handling price column
df['price']=df['price'].str.extract(r'(\d+.\d+)').astype(float)

# handling quantity column
df['book_quantity']=df['book_quantity'].str.extract(r'(\d+)').astype(int)


# handling seller_rate column
df['seller_rate']=df['seller_rate'].str.extract(r'(\d+)').astype(int)

# handling book_condition
df['book_condition']=df['book_condition'].str.replace('Condition: ','')

# splitting column location
df[['city','county','seller_country']]=df['seller_location'].str.split(',',n=2,expand=True)
df=df.drop(columns=['city','county'])


df['book_language']=df['book_language'].str.replace('Language: ','')

```

# DATA CLEANING ( 2_DETECTING MISSING AND DUPLICATES)

```python
for column in df.columns:
    print(f'column {column} has {df[column].isnull().sum()}')
print('\n',f'duplicates raw is {df.duplicated().sum()}')
```

# DATA CLEANING ( 2_HANDLING MISSING AND DUPLICATES)

```python
df=df.drop_duplicates()

df['shipping_price']=df['shipping_price'].fillna(df['shipping_price'].mean())

df['seller_country']=df['seller_country'].fillna(df['seller_country'].mode()[0])

df['Author']=df['Author'].fillna('Unknown')

df['published_year']=df['published_year'].fillna(df['published_year'].mean().astype(int))

df['published_year']=df['published_year'].astype(int)

# REMOVE NOT IMPORTANT COLUMN 
df=df.drop(columns=['book_language'])
```

# SAVING ALL CHANGES

```python
df.to_csv(r'C:\Users\Me\Downloads\DS_project_1\cleaned_abe_books.csv',index=False)
```
