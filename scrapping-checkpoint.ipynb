```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
book_title=[]
author_name=[]
book_language=[]
published_by=[]
published_year=[]
seller_name=[]
seller_location=[]
seller_rate=[]
case_book_type=[]
book_condition=[]
book_price=[]
shipping_price=[]
available_quantity=[]

page_num=0

while True:

    print(f'scrapping page {page_num}')
    
    url = f'https://www.abebooks.com/servlet/SearchResults?ds=30&dym=on&p={page_num}&rollup=on&sortby=20&sp=0&spo=30&tn=books'
    try:
        page = requests.get(url,timeout=10)
    except Exception as error:
        print (f"error occured \n: {error}")
    
    soup = BeautifulSoup(page.text,'lxml')
    main_div_lst = soup.find_all('div',class_='result-data col-xs-12 col-sm-9 cf')

    if not main_div_lst:
        print('no more books')
        break
    

    
    for div in main_div_lst:
    
        condition_1=div.find('h2',class_='title')
        book_title.append(condition_1.text.strip()
                         if condition_1 else None)
    
        condition_2=div.find('p',class_='author')
        author_name.append(condition_2.a.strong.text.strip()
                          if condition_2 else None)
    
        condition_3=div.find('p',class_='text-secondary')
        book_language.append(condition_3.text.strip()
                            if condition_3 else None)
    
        condition_4=div.find('span',class_='opt-publisher')
        published_by.append(condition_4.text.strip()
                           if condition_4 else None)
    
        condition_5=div.find('span',class_='opt-publish-date')
        published_year.append(condition_5.text.strip() 
                             if condition_5 else None)
    
        condition_6=div.find('span',{'data-test-id':'listing-seller-name'})
        seller_name.append(condition_6.text.strip()
                          if condition_6 else None)
    
        condition_7=div.find('span',{'data-test-id':'listing-seller-location'})
        seller_location.append(condition_7.text.strip()
                              if condition_7 else None)
    
        condition_8=div.find('span',class_='bookseller-rating')
        seller_rate.append(condition_8.text.strip()
                          if condition_8 else None)
    
        condition_9=div.find('span',{'data-test-id': 'listing-book-condition'})    
        case_book_type.append(condition_9.text.strip()
                             if condition_9 else None)
    
        condition_10=div.find('span',class_='opt-subcondition')
        book_condition.append(condition_10.text.strip()
                             if condition_10 else None)
    
        condition_11=div.find('p',class_='item-price')
        book_price.append(condition_11.text.strip() 
                         if condition_11 else None)

        condition_13=div.find('span',{'data-test-id':'shipping-detail'})
        shipping_price.append(condition_13.text.strip()
                             if condition_13 else None)
    
        condition_13=div.find('span',class_='opt-quantity')
        available_quantity.append(condition_13.text.strip()
                         if condition_13 else None)
    page_num=page_num+1
    time.sleep(3)

data={'Book_title':book_title,
      'price':book_price,
      'shipping_price':shipping_price,
      'book_quantity':available_quantity,
      'case_book_type':case_book_type,
      'book_condition':book_condition,
      'Author':author_name,
      'book_language':book_language,
      'published_by':published_by,
      'published_year':published_year,
      'seller_name':seller_name,
      'seller_location':seller_location,
      'seller_rate':seller_rate
     }    
df = pd.DataFrame(data)
df.to_csv('abe_books_web.csv',index=False)
```
